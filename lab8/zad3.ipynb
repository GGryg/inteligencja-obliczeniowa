```python
import snscrape.modules.twitter as gimmieTweets

query = "#pogrom"

tweets_list = []

for i, tweet in enumerate(gimmieTweets.TwitterSearchScraper(query).get_items()):
    if i > 100:
        break
    tweets_list.append(tweet)

for tweet in tweets_list:
    print(tweet.content)
```
