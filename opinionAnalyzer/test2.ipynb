```python
import pandas as pd

a1 = pd.read_csv('cv/nuclearenergy-Tweets-1-2023.csv')
a2 = pd.read_csv('cv/nuclearenergy-Tweets-2-2023.csv')
a3 = pd.read_csv('cv/nuclearenergy-Tweets-3-2023.csv')
a4 = pd.read_csv('cv/nuclearenergy-Tweets-4-2023.csv')
a5 = pd.read_csv('cv/nuclearenergy-Tweets-12-2022.csv')



acount = a1.shape[0] + a2.shape[0] + a3.shape[0] + a4.shape[0] + a5.shape[0]


all_tweets = a1['Text'].tolist() + a2['Text'].tolist() + a3['Text'].tolist() + a4['Text'].tolist() + a5['Text'].tolist()

print('isArt: ', acount)

allt = all_tweets[:4]
print(len(allt))
# 1513
# 2045
```

```python
import matplotlib.pyplot as plt
import text2emotion as te


def twice(allt):
    emotion_results = { "Happy": 0, "Angry": 0, "Surprise": 0, "Sad": 0, "Fear": 0 }
    emotion_tweets = []
    for tweet in allt:
        tweet_emotions = te.get_emotion(tweet)
        dominating_emotion = max(tweet_emotions, key=tweet_emotions.get)
        emotion_results[dominating_emotion] += 1
        
    numbers = [emotion_results["Happy"], emotion_results["Angry"], emotion_results["Surprise"], emotion_results["Sad"], emotion_results["Fear"]]
    labels = ['Happy', 'Angry', 'Surprise', 'Sad', 'Fear']

    plt.bar(labels, numbers)
    for i, num in enumerate(numbers):
        percentage = str(round(num/len(allt) * 100, 1)) + "%"
        plt.text(i, num, percentage, ha='center', va='bottom')

    plt.title('Text2Emotion Analysis')
    plt.xlabel('Emotion')
    plt.ylabel('Count')
    plt.show()

twice(allt)
twice(allt)
```
