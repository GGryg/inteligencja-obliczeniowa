```python
def how_many_blanks():
    count = 0
    for i in range(len(puzzle)):
        for j in range(len(puzzle[i])):
            if puzzle[i, j] == 0:
                count += 1
    return count

def make_2D_array(s):
    array_2D = []
    count = 0
    for i in range(len(puzzle)):
        inside_array = []
        for j in range(len(puzzle[i])):
            if puzzle[i, j] != 0 or count > len(s) - 1:
                inside_array.append(0)
            else:
                inside_array.append(round(s[count]))
                count += 1
        array_2D.append(inside_array)

    return array_2D
```

### Chromosomy
Chromosomy to cyfry od 1 do 9. Są długości ilości dostępnych miejsc do wstawienia cyfr na planszy.

### Funkcja fitness
Dostaje chromosomy i obraca je w 2 wymiarową tablicę.
Porównuje do planszy jak gdzieś znajdzie liczbę sumy to oblicza sumę z podanych cyfr dopóki nie znajdzie 0 w tablicy z chromosomami.
Jeśli suma nie jest taka sama jak liczba sumy, sprawdza czy jest to liczba sumy pozioma czy pionowa i zostaje obliczona suma cyfr jak blisko było do tej sumy i dodaje do fitness, który potem się minusuje.
Najlepszy fitness = 0.

```python
def fitness_func(solution, solution_idx):
    grid = make_2D_array(solution)
    col_diff = 0
    row_diff = 0

    for i in range(len(puzzle)):
        for j in range(len(puzzle[i])):
            if puzzle[i, j] > 0:
                sum_idx = puzzle[i, j]
                if puzzle_col_sums[sum_idx] != 0:
                    target_sum = puzzle_col_sums[sum_idx]
                    col_sum = 0
                    for i2 in range(i + 1, len(puzzle)):
                        if grid[i2][j] != 0:
                            col_sum += grid[i2][j]
                        else:
                            break
                    col_diff += abs(target_sum - col_sum)

                if puzzle_row_sums[sum_idx] != 0:
                    target_sum = puzzle_row_sums[sum_idx]
                    row_sum = 0
                    for j2 in range(j + 1, len(puzzle[j])):
                        if grid[i][j2] != 0:
                            row_sum += grid[i][j2]
                        else:
                            break
                    row_diff += abs(target_sum - row_sum)

    fitness = col_diff + row_diff
    return -fitness
```

### Ustawienia
1. Ilość genów zależy od ilości pustych miejsc na planszy
2. Około 50% populacji do "rozmnażania"
3. Ilość pokoleń: 1000
4. Selekcja rodziców steady
5. Krzyżowanie jednopunktowe
6. mutacja randomowa
7. Procent mutacji genów:
    1. Dla rozmiarów 4 x 4: 12
    2. Dla rozmiarów 9 x 8: 4
    3. Dla rozmiarów 15 x 15: 1
    
    Wyjaśnienie: procent mutacji genów * ilość dostępnych miejsc do wstawienia cyfr muszą dać blisko 100 ale conajmniej 100
8. Kryteria zatrzymania: Osiągnięcie fitness 0

```python
fitness_function = fitness_func

gene_space = [1, 2, 3, 4, 5, 6, 7, 8, 9]

sol_per_pop = 300
num_genes = how_many_blanks()

# rodzicow do "rozmanazania" 50% populacji
# 1000 pokoleń
# zachować 20% rodziców
num_parents_mating = 5
num_generations = 1000
keep_parents = 2

#typ seleckji rodzica: sss = steady
parent_selection_type = "sss"

crossover_type = "single_point"

mutation_type = "random"
# dla małych - 12
# dla średnich - 4
# dla dużych - 1
mutation_percent_genes = 1


ga_instance = pygad.GA(gene_space=gene_space,
                       num_generations=num_generations,
                       num_parents_mating=num_parents_mating,
                       fitness_func=fitness_function,
                       sol_per_pop=sol_per_pop,
                       num_genes=num_genes,
                       parent_selection_type=parent_selection_type,
                       keep_parents=keep_parents,
                       crossover_type=crossover_type,
                       mutation_type=mutation_type,
                       mutation_percent_genes=mutation_percent_genes,
                       stop_criteria="reach_0",
                       gene_type=int
                       )

start = time.time()
ga_instance.run()
end = time.time()

# podsumowanie: najlepsze znalezione czas, rozwiazanie chromosom, fitness i ile pokoleń mineło
t = end - start
print("time: ", t)

solution, solution_fitness, solution_idx = ga_instance.best_solution()
print("Parameters of the best solution : {solution}".format(solution=solution))
print("Fitness value of the best solution = {solution_fitness}".format(solution_fitness=solution_fitness))
print("Generations passed: {generations_completed}".format(generations_completed=ga_instance.generations_completed))


# wykres
ga_instance.plot_fitness()
```

```python
from pyswarms.single.global_best import GlobalBestPSO

g_size = how_many_blanks()

print(puzzle_row_sums)

def scores(x):
    grid = make_2D_array(x)
    score = 0

    col_diff = 0
    row_diff = 0

    for i in range(len(puzzle)):
        for j in range(len(puzzle[i])):
            if puzzle[i, j] > 0:
                sum_idx = puzzle[i, j]
                if puzzle_col_sums[sum_idx] != 0:
                    target_sum = puzzle_col_sums[sum_idx]
                    col_sum = 0
                    for i2 in range(i + 1, len(puzzle)):
                        if grid[i2][j] != 0:
                            col_sum += grid[i2][j]
                        else:
                            break
                    col_diff += abs(target_sum - col_sum)

                if puzzle_row_sums[sum_idx] != 0:
                    target_sum = puzzle_row_sums[sum_idx]
                    row_sum = 0
                    for j2 in range(j + 1, len(puzzle[j])):
                        if grid[i][j2] != 0:
                            row_sum += grid[i][j2]
                        else:
                            break
                    row_diff += abs(target_sum - row_sum)

    score = col_diff + row_diff
    return score

def f(x):
    j = [scores(x[i]) for i in range(len(x))]
    return np.array(j)

bounds = (np.ones(g_size), np.ones(g_size)*9)

options = {'c1': 0.5, 'c2': 0.5, 'w':0.7}

optimizer = GlobalBestPSO(n_particles=400, dimensions=g_size, bounds=bounds, options=options)

cost, solution = optimizer.optimize(f, iters=100)
print(make_2D_array(solution))
```
