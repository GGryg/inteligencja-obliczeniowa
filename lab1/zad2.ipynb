Zad 2 a
Wczytanie pliku csv i wyświetlenie wartości

```python
import pandas
import matplotlib.pyplot as plt

miasta = pandas.read_csv("miasta.csv")

print(miasta)

print(miasta.values)
```

Zad2 b
Dodanie wartości

```python
df = pandas.DataFrame([[2010, 460, 555, 405]], columns=('Rok', 'Gdansk', 'Poznan', 'Szczecin'))
miasta = miasta.append(df, ignore_index=True)
miasta = miasta.set_index('Rok')

print(miasta)
```

Zad2 c
Utworzenie wykresu dla ludności Gdańska

```python
plt.plot(miasta['Gdansk'], marker='o', color='r')
plt.title('Ludność Gdańska')
plt.ylabel('Liczba ludności [w tys.]')
#plt.xticks(range(len(miasta['Rok'].values)), miasta['Rok'].values)
plt.xlabel('Lata')
```

Zad2 d
Utworzenie wykresu dla Ludności Polski

```python
fig, ax = plt.subplots()
ax.plot(miasta['Gdansk'], 'r-o', miasta['Poznan'], 'g-o', miasta['Szczecin'], 'b-o')
plt.legend(['Gdansk', 'Poznan', 'Szczecin'])
plt.title('Ludnosc w Polsce')
plt.ylabel('Liczba ludnosci [w tys.]')
#plt.xticks(range(len(miasta['Rok'].values)), miasta['Rok'].values)
plt.xlabel('lata')
```

Zad2 e
Przeskalowanie danych standaryzacją oraz obliczenie średniej i odchylenia standardowego

```python
standard = (miasta - miasta.mean()) / miasta.std()
print(standard)
print("Średnia ", standard.mean())
print("Odchylenie standardowe ", standard.std())
```

Zad2 f
Przeskalowanie danych normalizacją oraz obliczenie wartości minimalnej i maksymalnej

```python
normal = (miasta - miasta.min()) / (miasta.max() - miasta.min())
print(normal)
print("Min ", normal.min())
print("Max", normal.max())
```
